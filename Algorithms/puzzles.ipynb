```python
#  100 door challenge

doors = [False] * 101

for i in range(1,101):
    for j in range(i,101,i):
        doors[j] = not doors[j]
        
for i in range(1,101):
    if doors[i] is True:
        print(i)
```

```python
#  fizz buzz

value = list(range(1,101))
for i in value:
    if (i % 5 == 0) and (i % 3 == 0):
        print("fizz buzz") 
    elif i % 3 == 0:
        print("fizz")
    elif i % 5 == 0:
        print("buzz")
      
    else:
        print(i)
    
```
